## Data preprocessing

## Feature engineering

```python
# year birth can be converted to age 
# 2026 is a reference date here
df["Age"]=2026-df["Year_Birth"]
df.head()
```

```python
# converting join date to costomer tenure days 
# reference date = last customer join date 
# for ex reference date is 5 and and another 3 then the tenure of another customer is 5-3=2 days
```

```python
# customer joining date
df["Dt_Customer"]=pd.to_datetime(df["Dt_Customer"],dayfirst=True)
reference_date=df["Dt_Customer"].max()
df["Customer_tenure_days"]=(reference_date-df["Dt_Customer"]).dt.days
```

## Encoding

## Scaling

```python
X=df_encoded
scaler=StandardScaler()
x_scaled=scaler.fit_transform(X)
```

## Visualising the data

```python
# here are 18 dimensions so we cant visualize the data so we will use PCA
# for ex:
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)
X_pca=pca.fit_transform(x_scaled)
```

```python
# plot
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection="3d")

ax.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2])
ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
ax.set_title("3d projection")


```

```python
pca.explained_variance_ratio_
```

## Analysing k value

1. Elbow method

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit_predict(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
knee=KneeLocator(range(1,11),wcss,curve="convex",direction="decreasing")
optimal_k=knee.elbow
```

```python
print("best k =",optimal_k)
```

```python
plt.plot(range(1,11),wcss,marker="o")
plt.xlabel("k")
plt.ylabel("wcss")
```

2.Silhouette Score

```python
from sklearn.metrics import silhouette_score

scores=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k, random_state=42)
    labels=kmeans.fit_predict(X_pca)
    score=silhouette_score(X_pca,labels)
    scores.append(score)

# plot 
plt.plot(range(2,11),scores,marker="o")
plt.xlabel("k")
plt.ylabel("Silhoette")
```

```python
k_range=range(2,11)

fig,ax1 =plt.subplots(figsize=(8,6))

ax1.plot(k_range,wcss[:len(k_range)],marker="o",color="blue")
ax1.set_xlabel("k")
ax1.set_ylabel("wcss")

ax2=ax1.twinx()
ax2.plot(k_range,scores[:len(k_range)],marker="x",color="red",linestyle="--")
ax2.set_ylabel("SS")
```

## Clustering

1.Kmeans

```python
kmeans=KMeans(
    n_clusters=4,
    random_state=42
)
labels_kmeans=kmeans.fit_predict(X_pca)
```

```python
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection="3d")

ax.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2],c=labels_kmeans)
```

2.Agglomarative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg_clf=AgglomerativeClustering(
    n_clusters=4,
    linkage="ward"
)
labels_agg=agg_clf.fit_predict(X_pca)
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection="3d")
ax.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2],c=labels_agg)
```

## Agglomarative is better than kmeans
